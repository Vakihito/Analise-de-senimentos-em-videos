```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sift_data = pd.read_pickle('sift_gray_p.pkl')
sift_data.head(1)
```

```python
def position_2d_to_1d(position):
  return position[:,0] * 200 + position[:,1]
```

```python
list_arr_points = []
for cur_sift_p in sift_data['sift_p']:
  list_arr_points.append(np.array(cur_sift_p).astype(int)) 
```

```python
sift_data['sift_p_arr'] = list_arr_points
```

```python
sift_data['sift_p_arr_f'] = sift_data['sift_p_arr'].apply(lambda x: position_2d_to_1d(x))
```

```python
sift_data['sift_p_arr_f_size'] = sift_data['sift_p_arr_f'].apply(lambda x: len(x))
```

```python
sift_data.head(1)
```

```python
def padd_array(cur_arr, max_size):
  return np.pad(cur_arr, (0, max_size - len(cur_arr)) ,'constant', constant_values=(-1))
```

```python
max_size =  np.max(sift_data['sift_p_arr_f_size'].values) + 1
max_size
```

```python
sift_data['sift_p_arr_f_p'] = sift_data['sift_p_arr_f'].apply(lambda x: padd_array(x,max_size))
```

```python
sift_data.head(1)
```

```python
sift_data_inp = sift_data[['img_location', 'x0', 'y0', 'x1', 'y1', 'sift_p_arr_f_p', 'sift_p_arr_f_size']].copy()
```

```python
sift_data_inp['sift_p_arr_f_p'][0]
```

```python
def normalize_arr(arr, upper=512):
  diff = arr.max() - arr.min()
  min_v = arr.min()
  arr_norm = (arr - min_v) * (upper/diff)
  return arr_norm.astype(int)
```

```python
sift_data_inp['sift_p_arr_f_p_q'] = sift_data_inp['sift_p_arr_f_p'].apply(lambda x: normalize_arr(x))
```

```python
sift_data_inp.head(1)
```

```python
sift_data_inp['sift_p_arr_hist'] = sift_data_inp['sift_p_arr_f_p_q'].apply(lambda x: np.histogram(x, bins=512, range=(1,512))[0])
```
